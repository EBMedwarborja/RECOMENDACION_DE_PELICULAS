```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pyarrow.parquet as pq
import seaborn as sns
import pandas as pd
%load_ext autoreload
%autoreload 2
import warnings
warnings.filterwarnings("ignore")

# Cargar los archivos parquet
movies_dataset = pd.read_parquet(r'C:\Users\edwar\Documents\PROYETO_DTA_01_SISTEMA_DE_RECOMENDACION\DATASET\movies_dataset.parquet')
cast_data = pd.read_parquet(r'C:\Users\edwar\Documents\PROYETO_DTA_01_SISTEMA_DE_RECOMENDACION\DATASET\cast_data.parquet')
crew_data = pd.read_parquet(r'C:\Users\edwar\Documents\PROYETO_DTA_01_SISTEMA_DE_RECOMENDACION\DATASET\crew_data.parquet')
```




Para realizar un análisis exploratorio de los datos (Exploratory Data Analysis - EDA), vamos a analizar las variables clave del dataset, buscar patrones, identificar posibles outliers y explorar la frecuencia de palabras en los títulos de películas. Este proceso incluye visualizaciones y cálculos estadísticos que permiten descubrir insights valiosos.
Pasos para el EDA:

    Descripción general de las variables:
        Identificar las columnas y los tipos de datos.
        Resumen estadístico de las variables numéricas.

    Distribución de variables numéricas:
        Crear histogramas y boxplots para analizar la distribución de variables como revenue, budget, y popularity.
        Identificar posibles outliers o anomalías.

    Análisis de correlaciones:
        Calcular la correlación entre variables numéricas para ver si existen relaciones fuertes entre ellas.

    Distribución temporal:
        Análisis de la distribución de películas a lo largo del tiempo, usando release_year.

    Frecuencia de palabras en los títulos:
        Crear una nube de palabras para analizar las palabras más frecuentes en los títulos de las películas.

    Visualización de outliers:
        Boxplots para detectar outliers en variables clave como revenue, budget, y popularity.




```python
# Verificar que los datos se han cargado correctamente
print("Movies Dataset:")
print(movies_dataset.head())

print("\nCast Data:")
print(cast_data.head())

print("\nCrew Data:")
print(crew_data.head())
```

```python
# Resumen estadístico de las variables numéricas
print(movies_dataset.describe())
print(cast_data.describe())
print(crew_data.describe())
# Información del dataset para verificar tipos de datos y valores nulos
print(movies_dataset.info())
print(cast_data.info())
print(crew_data.info())
```

```python
# Analizar la cantidad de valores nulos por columna
missing_values = movies_dataset.isnull().sum()
print(missing_values[missing_values > 0])

# Visualizar los valores nulos con un heatmap
plt.figure(figsize=(12,8))
sns.heatmap(movies_dataset.isnull(), cbar=False, cmap='viridis')
plt.title('Valores Nulos en el Dataset')
plt.show()
```

```python
# Histograma y KDE para popularidad
plt.figure(figsize=(10,5))
sns.histplot(movies_dataset['popularity'], bins=50, kde=True, color='blue')
plt.title('Distribución de Popularidad')
plt.show()

# Histograma y KDE para budget
plt.figure(figsize=(10,5))
sns.histplot(movies_dataset['budget'], bins=50, kde=True, color='green')
plt.title('Distribución de Presupuesto (Budget)')
plt.show()

# Histograma y KDE para revenue
plt.figure(figsize=(10,5))
sns.histplot(movies_dataset['revenue'], bins=50, kde=True, color='red')
plt.title('Distribución de Ingresos (Revenue)')
plt.show()

# Histograma y KDE para return
plt.figure(figsize=(10,5))
sns.histplot(movies_dataset['return'], bins=50, kde=True, color='purple')
plt.title('Distribución de Retorno (Return)')
plt.show()
```

```python
# Boxplot para revenue, budget y return
plt.figure(figsize=(12,6))
sns.boxplot(data=movies_dataset[['revenue', 'budget']], palette='Set3')
plt.title('Boxplot de Revenue y Budget ')
plt.show()
```

```python
# Relación entre Budget y Revenue
plt.figure(figsize=(8,6))
sns.scatterplot(x='budget', y='revenue', data=movies_dataset)
plt.title('Relación entre Presupuesto y Ingresos')
plt.show()

```

```python
# Conteo de actores más frecuentes
top_actors = cast_data['name'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_actors.values, y=top_actors.index, palette='coolwarm')
plt.title('Actores más Frecuentes en las Películas')
plt.show()

```

```python

# Distribución del género en cast_data
plt.figure(figsize=(8, 5))
sns.countplot(data=cast_data, x='gender')
plt.title('Distribución de Género en los Actores')
plt.xlabel('Género (0: No especificado, 1: Mujer, 2: Hombre)')
plt.ylabel('Frecuencia')
plt.show()

# Distribución del género en crew_data
plt.figure(figsize=(8, 5))
sns.countplot(data=crew_data, x='gender')
plt.title('Distribución de Género en el Equipo')
plt.xlabel('Género (0: No especificado, 1: Mujer, 2: Hombre)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Filtrar directores en crew_data
directors_data = crew_data[crew_data['job'] == 'Director']

# Contar el número de películas dirigidas por cada director
director_counts = directors_data['name'].value_counts().head(10)

# Visualizar los directores más productivos
plt.figure(figsize=(10, 5))
sns.barplot(x=director_counts.index, y=director_counts.values)
plt.title('Top 10 Directores más Productivos')
plt.xlabel('Director')
plt.ylabel('Número de Películas')
plt.xticks(rotation=45)
plt.show()
```

```python
# Contar el número de trabajos por departamento
department_counts = crew_data['department'].value_counts()

# Visualizar la distribución de trabajos por departamento
plt.figure(figsize=(10, 5))
sns.barplot(x=department_counts.index, y=department_counts.values)
plt.title('Distribución de Trabajos por Departamento en el Equipo')
plt.xlabel('Departamento')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.show()
```

```python
# Asegurarse de que la columna 'id' en ambos DataFrames tenga el mismo tipo
cast_data['id'] = cast_data['id'].astype(str)
movies_dataset['id'] = movies_dataset['id'].astype(str)

# Unir cast_data con movies_dataset en base a la película
cast_with_revenue = pd.merge(cast_data, movies_dataset[['id', 'return']], left_on='id', right_on='id', how='left')

# Calcular el promedio de retorno de inversión por actor
actor_success = cast_with_revenue.groupby('name')['return'].mean().sort_values(ascending=False).head(10)

# Visualizar los actores con mayor retorno de inversión
plt.figure(figsize=(10, 5))
sns.barplot(x=actor_success.index, y=actor_success.values)
plt.title('Top 10 Actores con Mayor Retorno de Inversión')
plt.xlabel('Actor')
plt.ylabel('Promedio de Retorno de Inversión')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualizar la distribución del retorno de inversión
plt.figure(figsize=(10,5))
sns.histplot(movies_dataset['return'], bins=50, kde=True)
plt.title('Distribución del Retorno de Inversión de las Películas')
plt.xlabel('Retorno de Inversión')
plt.ylabel('Frecuencia')
plt.xlim(0, 10)  # Limitar para excluir valores extremadamente altos
plt.show()

# Películas con mayor retorno de inversión
top_return_movies = movies_dataset[['title', 'return']].sort_values(by='return', ascending=False).head(10)
print("Películas con mayor retorno de inversión:")
print(top_return_movies)
```

```python
# Visualizar la distribución de la popularidad
plt.figure(figsize=(10,5))
sns.histplot(movies_dataset['popularity'], bins=50, kde=True)
plt.title('Distribución de la Popularidad de las Películas')
plt.xlabel('Popularidad')
plt.ylabel('Frecuencia')
plt.show()

# Películas más populares
top_popular_movies = movies_dataset[['title', 'popularity']].sort_values(by='popularity', ascending=False).head(10)
print("Películas más populares:")
print(top_popular_movies)
```

```python
# Visualizar la distribución de los presupuestos
plt.figure(figsize=(10,5))
sns.histplot(movies_dataset['budget'], bins=50, kde=True)
plt.title('Distribución del Presupuesto de las Películas')
plt.xlabel('Presupuesto')
plt.ylabel('Frecuencia')
plt.xlim(0, movies_dataset['budget'].quantile(0.95))  # Limitar al 95% para excluir valores extremos
plt.show()

# Películas con mayor presupuesto
top_budget_movies = movies_dataset[['title', 'budget']].sort_values(by='budget', ascending=False).head(10)
print("Películas con mayor presupuesto:")
print(top_budget_movies)
```

```python
# Relación entre presupuesto y retorno de inversión
plt.figure(figsize=(10,5))
sns.scatterplot(x='budget', y='return', data=movies_dataset)
plt.title('Relación entre Presupuesto y Retorno de Inversión')
plt.xlabel('Presupuesto')
plt.ylabel('Retorno de Inversión')
plt.xlim(0, movies_dataset['budget'].quantile(0.95))  # Limitar para excluir valores extremos
plt.ylim(0, 10)  # Limitar para excluir valores extremadamente altos
plt.show()
```

```python
# Convertir las columnas 'return' y 'popularity' a numérico, forzando errores a NaN
movies_dataset['return'] = pd.to_numeric(movies_dataset['return'], errors='coerce')
movies_dataset['popularity'] = pd.to_numeric(movies_dataset['popularity'], errors='coerce')

# Reemplazar NaN con 0 (opcional, dependiendo de lo que quieras hacer con los NaN)
movies_dataset['return'].fillna(0, inplace=True)
movies_dataset['popularity'].fillna(0, inplace=True)

# Calcular el éxito como el producto de 'return' y 'popularity'
movies_dataset['success'] = movies_dataset['return'] * movies_dataset['popularity']

# Mostrar las películas más exitosas combinando retorno y popularidad
top_success_movies = movies_dataset[['title', 'success']].sort_values(by='success', ascending=False).head(10)
print("Películas más exitosas combinando retorno y popularidad:")
print(top_success_movies)
```
